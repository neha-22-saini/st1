1. Random Sampling
•	Definition: Every individual (or unit) in the population has an equal chance of being selected.
•	Process: Selection is completely by chance (like picking names from a hat or using a random number generator).
•	Use case: Works best when the population is homogeneous (similar characteristics).
•	Advantage: Simple and unbiased if the population is well mixed.
•	Disadvantage: If the population is diverse/heterogeneous, some groups may be underrepresented or not represented at all.
2. Stratified Sampling
•	Definition: The population is divided into strata (subgroups) based on certain characteristics (e.g., age, gender, income), and then samples are randomly chosen from each stratum.
•	Process: Two steps:
1.	Divide population into relevant strata.
2.	Perform random sampling within each stratum (proportional to its size, usually).
•	Use case: Best when the population has clear subgroups with different characteristics.
•	Advantage: Ensures representation of all groups, increases accuracy and reduces sampling error.
•	Disadvantage: More complex to organize, requires prior knowledge about the population.

Question 3. Define mean, median, and mode. Explain why these measures of central tendency are important.

Answer.
            1. Mean (Arithmetic Average)
•	Definition: The sum of all values in a dataset divided by the number of values.
                 Mean=Number of observations/Sum of all observations 

•	Example: For 2, 4, 6, 8 → 
                           Mean = (2+4+6+8) /4 = 5

2. Median
•	Definition: The middle value of a dataset when the values are arranged in order.
•	If the number of values is odd → the middle one is the median.
•	If even → median is the average of the two middle values.
•	Example: For 1, 3, 5, 7, 9 → Median = 5
For 2, 4, 6, 8 → Median = (4+6)/2 = 5

3. Mode
•	Definition: The value that appears most frequently in a dataset.
•	A dataset can have:
o	One mode (unimodal)
o	More than one mode (bimodal/multimodal)
o	No mode (if all values occur only once)
•	Example: For 2, 3, 3, 4, 5 → Mode = 3

why these measures of central tendency are important??
-->
1. Simplify Large Data
•	Instead of dealing with hundreds or thousands of numbers, these measures give one representative value that summarizes the data.
2. Allow Comparisons
•	They make it easier to compare different groups or datasets.
3. Identify Patterns
•	They reveal the typical or common outcome in the data.
4. Handle Skewed Data
•	When data has extreme values (like income), the median gives a fairer picture than the mean
5. Basis for Further Analysis
•	Many statistical methods (like variance, standard deviation, hypothesis testing) are built on the mean, median, or mode.


Question 4) Explain skewness and kurtosis. What does a positive skew imply about the data?
Ans.
     1. Skewness
•	Definition: Skewness measures the asymmetry of a distribution.
•	A perfectly symmetrical dataset (like a normal distribution) has skewness = 0.
Types:
1.	Positive Skew (Right-skewed)
o	Tail is stretched to the right (towards larger values).
o	Mean > Median > Mode.
o	Example: Income distribution (a few people earn extremely high amounts).
2.	Negative Skew (Left-skewed)
o	Tail is stretched to the left (towards smaller values).
o	Mean < Median < Mode.
o	Example: Age at retirement (most people retire around the same age, but a few retire much earlier)



2. Kurtosis
•	Definition: Kurtosis measures the "tailedness" or how peaked/flat a distribution is compared to a normal distribution.
Types:
1.	Mesokurtic (kurtosis ≈ 3) → Normal distribution.
2.	Leptokurtic (kurtosis > 3) → Heavy tails, sharp peak, more outliers.
3.	Platykurtic (kurtosis < 3) → Light tails, flatter peak, fewer outliers.

What does a Positive Skew imply?
•	The data has a long tail on the right side.
•	Most values are concentrated on the left (lower end), but a few large extreme values pull the mean to the right.
•	Typical interpretation:
o	Mean > Median > Mode
o	Data is not balanced, and average may not represent the “typical” case.
o	Common in income, housing prices, waiting times, etc.



Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers. 
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

```python

    
import statistics as stats           # importing statistics 

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Compute mean, median, and mode
mean_val = stats.mean(numbers)
median_val = stats.median(numbers)
mode_val = stats.mode(numbers)

# Print results
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```


Question 6 : Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60]
Ans. 


```python
import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to numpy arrays
x = np.array(list_x)
y = np.array(list_y)

# Compute covariance matrix
cov_matrix = np.cov(x, y, bias=False)   # bias=False → sample covariance
cov_xy = cov_matrix[0, 1]              # covariance between x and y

# Compute correlation coefficient
corr_xy = np.corrcoef(x, y)[0, 1]

# Print results
print("Covariance between x and y:", cov_xy)
print("Correlation coefficient between x and y:", corr_xy)
```

Question: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


Ans.


```python
import matplotlib.pyplot as plt

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw a boxplot
plt.boxplot(data, vert=False, patch_artist=True)
plt.title("Boxplot of Data")
plt.xlabel("Values")
plt.show()
```

How Outliers are Detected (IQR Method):
•	Q1 (25th percentile) = 18
•	Q3 (75th percentile) = 24
•	IQR = Q3 - Q1 = 6
•	Lower bound = Q1 - 1.5×IQR = 18 - 9 = 9
•	Upper bound = Q3 + 1.5×IQR = 24 + 9 = 33
 Any value below 9 or above 33 is an outlier.

 Outlier in the Data:
•	The value 35 lies above 33 → it is an outlier.
•	All other values are within the normal range.

 Explanation of Result:
•	The boxplot will show:
o	Box → between Q1 (18) and Q3 (24).
o	Median line around 22.
o	Whiskers extending to min (12) and max (29).
o	35 will be plotted separately as a dot, indicating an outlier.



Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales. 
● Explain how you would use covariance and correlation to explore this relationship. 
● Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000]

Ans. 
How to use Covariance and Correlation
•	Covariance
o	Tells whether two variables move together (positive covariance) or in opposite directions (negative covariance).
o	Example: If advertising spend increases and sales also increase, covariance will be positive.
o	Limitation: Covariance does not tell us the strength of the relationship, only the direction.
•	Correlation
o	Standardizes covariance (scales it between -1 and +1).
o	+1 → perfect positive relationship
o	-1 → perfect negative relationship
o	0 → no relationship
o	In this case, correlation will tell us how strongly advertising spend is related to sales.
 So, covariance tells the direction, and correlation tells the direction + strength.


```python
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to numpy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Covariance matrix
cov_matrix = np.cov(x, y, bias=False)
cov_xy = cov_matrix[0, 1]

# Correlation coefficient
corr_xy = np.corrcoef(x, y)[0, 1]

# Print results
print("Covariance:", cov_xy)
print("Correlation Coefficient:", corr_xy)
```
